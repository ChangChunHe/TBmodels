```python
latt = kwant.lattice.general(
    prim_vecs=[[1, 0],[0,0.5]],
    basis=[[0,0],[0.5,0.5]]
)
```

```python
sys = kwant.Builder()
for i in range(5):
    for j in range(4):
        for l1 in latt.sublattices:
            for l2 in latt.sublattices:
                sys[l1(i,j)] = 0.1
                sys[l1(i,j)] += 0.1
                if j > 0:
                    sys[l1(i, j), l2(i, j - 1)] = 0.2
```

```python
kwant.plot(sys)
```
