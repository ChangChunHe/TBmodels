```python
latt = kwant.lattice.general(
    prim_vecs=[[1, 0],[0,0.5]],
    basis=[[0,0],[0.5,0.5]]
)
```

```python
sys = kwant.Builder()
for i in range(5):
    for j in range(4):
        for k, l in enumerate(latt.sublattices):
            sys[l(i,j)]= k + 0.1
            if i > 0:
                sys[l(i, j), l(i - 1, j)] = 0.1
            if j > 0:
                sys[l(i,j), l(i, j - 1)] = 0.2
```

```python
kwant.plot(sys)
```
